```python

df1 = pd.read_csv("trafik_verisi/accidents_2005_to_2007.csv", low_memory=False)
df2 = pd.read_csv("trafik_verisi/accidents_2009_to_2011.csv", low_memory=False)
df3 = pd.read_csv("trafik_verisi/accidents_2012_to_2014.csv", low_memory=False)

df = pd.concat([df1, df2, df3], ignore_index=True)
df.head()
```

<h2>İstatistiksel Özet – Merkezi Eğilim & Dağılım Ölçüleri</h2>

```python
numerical_cols = df.select_dtypes(include='number').columns

# describe ile temel istatistiksel özet çıkardım
summary = df[numerical_cols].describe().T
```

```python
summary["median"] = df[numerical_cols].median()
summary["missing"] = df[numerical_cols].isnull().sum()
summary["missing_ratio"] = (df[numerical_cols].isnull().mean() * 100).round(2)
summary["n_unique"] = df[numerical_cols].nunique()
```

```python
print(" Sayısal Değişkenlerin İstatistiksel Özeti:")
summary.sort_index()
```

<h2>Eksik Değer Analizi</h2>

```python
# Eksik değerleri saydırdım
missing = df.isnull().sum()
```

```python
# Sadece eksik olanları filtreledim
missing = missing[missing > 0].sort_values(ascending=False)
```

```python
# Oran hesapladım
missing_percent = (missing / len(df) * 100).round(2)
```

```python
# Sonuçları DataFrameE çevirdim
missing_df = pd.DataFrame({
    "Eksik Değer Sayısı": missing,
    "Yüzde (%)": missing_percent
})
```

```python
print(" Eksik Değer Özeti:")
missing_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.rcParams['font.family'] = 'DejaVu Sans'

plt.figure(figsize=(10, 6))
sns.barplot(x=missing.values, y=missing.index, palette="mako")
plt.title("Eksik Değer Dağılımı")
plt.xlabel("Eksik Değer Sayısı")
plt.ylabel("Değişken")
plt.tight_layout()
plt.show()
```

<h2> Veri Temizleme </h2>

```python
df.drop(columns=["Junction_Detail", "Carriageway_Hazards", "Special_Conditions_at_Site"], inplace=True)
```

```python
fill_mode_cols = [
    "Weather_Conditions", "Road_Surface_Conditions", "Light_Conditions",
    "Pedestrian_Crossing-Physical_Facilities", "Pedestrian_Crossing-Human_Control"
]

for col in fill_mode_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.dropna(subset=["Latitude", "Longitude"], inplace=True)
```

```python
# eksik veri kaldı mı kontrol
print("Kalan eksik değerler:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

<h2>Aykırı Değer (Outlier) Analizi</h2>

```python
#Kullanacağımız Yöntem: IQR (Interquartile Range)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sayısal kolonları belirledim
target_cols = ["Number_of_Vehicles", "Number_of_Casualties", "Speed_limit"]
```

```python
#Aykırı değerleri analiz ettim
for col in target_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_ratio = len(outliers) / len(df) * 100

    print(f"\n {col} sütunu:")
    print(f"Alt sınır: {lower_bound}, Üst sınır: {upper_bound}")
    print(f"Aykırı değer sayısı: {len(outliers)} ({outlier_ratio:.2f}%)")
```

```python
# Boxplot ile gösterdim
plt.figure(figsize=(8, 1.5))
sns.boxplot(x=df[col], color='tomato')
plt.title(f"{col} - Aykırı Değer Görselleştirme")
plt.show()
```

<h2>Görselleştirme Analizi</h2>

```python
# Saat, yıl, gün gibi sütunları ürettim
df["Date"] = pd.to_datetime(df["Date"], errors="coerce", dayfirst=True)
df["Hour"] = pd.to_datetime(df["Time"], errors="coerce").dt.hour
df["Year"] = df["Date"].dt.year
df["Day_of_Week"] = df["Day_of_Week"].map({
    1: "Pazar", 2: "Pazartesi", 3: "Salı", 4: "Çarşamba",
    5: "Perşembe", 6: "Cuma", 7: "Cumartesi"
})
df["Accident_Severity"] = df["Accident_Severity"].map({1: "Fatal", 2: "Serious", 3: "Slight"})
```

```python
#Saatlik kaza dağılımı
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="Hour", palette="crest")
plt.title(" Saatlik Kaza Dağılımı")
plt.xlabel("Saat")
plt.ylabel("Kaza Sayısı")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Haftanın günlere göre dağılımı
plt.figure(figsize=(9, 5))
order = ["Pazartesi", "Salı", "Çarşamba", "Perşembe", "Cuma", "Cumartesi", "Pazar"]
sns.countplot(data=df, x="Day_of_Week", order=order, palette="flare")
plt.title(" Haftanın Günlerine Göre Kaza Dağılımı")
plt.xlabel("Gün")
plt.ylabel("Kaza Sayısı")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Kaza Şiddetine Göre Dağılım
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Accident_Severity", palette="rocket")
plt.title(" Kaza Şiddeti Dağılımı")
plt.xlabel("Şiddet")
plt.ylabel("Kaza Sayısı")
plt.tight_layout()
plt.show()
```
